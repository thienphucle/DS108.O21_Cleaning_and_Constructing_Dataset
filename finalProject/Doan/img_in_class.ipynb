```python
import os
from  matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
total_num = 0
targetnames = ['akiec', 'bcc', 'bkl', 'df', 'mel', 'nv', 'vasc']

temp = []
for cla in targetnames:
    cla_path = os.path.join(fr"dataset_ds108\train_dir", cla)
    images = os.listdir(cla_path)
    num = len(images)
    total_num += num
    temp.append(num)
    
bar, ax = plt.subplots(figsize = (5,5))
plt.bar(targetnames, temp, width = 0.4)
```

```python
total_num = 0

temp = []
for cla in targetnames:
    cla_path = os.path.join(fr"dataset_ds108\train_4_hr", cla)
    images = os.listdir(cla_path)
    num = len(images)
    total_num += num
    temp.append(num)
    
bar, ax = plt.subplots(figsize = (5,5))
plt.bar(targetnames, temp, width = 0.4)
```

```python
total_num = 0

temp = []
for cla in targetnames:
    cla_path = os.path.join(fr"dataset_ds108\test_dir_4_hr", cla)
    images = os.listdir(cla_path)
    num = len(images)
    total_num += num
    temp.append(num)
    
bar, ax = plt.subplots(figsize = (5,5))
plt.pie(temp, labels = targetnames, autopct="%.1f%%")
plt.title('Images in test dir', size=16)
```
